## Negative Transformation of Image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# image = cv2.imread('../5.3.01.jpg')
image = cv2.imread('../sceneary.jpg')
```

```python
image
```

```python
image.shape
```

```python
plt.imshow(image);
```

```python
color = ('b', 'g', 'r')

for i, col in enumerate(color):
    histr = cv2.calcHist(images=image, channels=[i], mask=None, histSize=[256], ranges=[0, 256])
    plt.plot(histr, color = col)
    plt.xlim([0, 256])
plt.show()
```

Taking negative of image

```python
negative_image = 255 - image
# negative_image = 1 - image
```

```python
plt.imshow(negative_image);
```

```python
color = ('b', 'g', 'r')

for i, col in enumerate(color):
    histr = cv2.calcHist(images=negative_image, channels=[i], mask=None, histSize=[256], ranges=[0, 256])
    plt.plot(histr, color = col)
    plt.xlim([0, 256])

plt.show()
```

```python
cv2.imwrite('negative_image.png',negative_image)
```
