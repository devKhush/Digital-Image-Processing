```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import matplotlib as mpl
mpl.rcParams['figure.facecolor'] = 'white'
```

```python
# implement linear interpolation
def linear_interpolation(img, scale):
    h, w, c = img.shape
    new_h = int(h * scale)
    new_w = int(w * scale)
    new_img = np.zeros((new_h, new_w, c), dtype=np.uint8)
    for i in range(new_h):
        for j in range(new_w):
            x = i / scale
            y = j / scale
            x1 = int(x)
            y1 = int(y)
            x2 = min(x1 + 1, h - 1)
            y2 = min(y1 + 1, w - 1)
            new_img[i, j] = (x2 - x) * (y2 - y) * img[x1, y1] + (x2 - x) * (y - y1) * img[x1, y2] + (x - x1) * (y2 - y) * img[x2, y1] + (x - x1) * (y - y1) * img[x2, y2]
    return new_img
```

```python
image = cv2.imread('../../images/sceneary.png')
plt.imshow(image);
```

```python
interpolated_img = linear_interpolation(image, 2);
```

```python
plt.imsave('linear-interpolation.png', interpolated_img)
plt.imshow(interpolated_img);
```
