```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.facecolor'] = 'white'
```

```python
image = cv2.imread('../../images/sceneary.png')
plt.imshow(image);
```

```python
# implement nearest neighbour interpolation
def nearest_neighbour_interpolation(img, scale):
    # get the shape of the image
    h, w, c = img.shape
    # get the shape of the new image
    new_h = int(h * scale)
    new_w = int(w * scale)
    # create the new image
    
    new_img = np.zeros((new_h, new_w, c), dtype=np.uint8)
    # fill the new image
    for i in range(new_h):
        for j in range(new_w):
            # get the nearest neighbour
            x = int(i / scale)
            y = int(j / scale)
            # fill the new image
            new_img[i, j] = img[x, y]
    return new_img
```

```python
interpolated_img = nearest_neighbour_interpolation(image, 2)
plt.imshow(interpolated_img)
```

```python
cv2.imwrite('interpolated_image.png', interpolated_img)
```
