```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

### Contrasting a Image 

```python
image = cv2.imread('../images/table.jpg')
```

```python
plt.imshow(image);
```

```python
def contrast_streching(r1, s1, r2, s2, pixel_value):
    if 0 <= pixel_value and pixel_value < r1:
        return pixel_value * (s1/r1)
    elif r1 <= pixel_value and pixel_value < r2:
        return ((s2 - s1)/(r2 - r1)) * (pixel_value - r1) + s1
    else:
        return ((255 - s2)/(255 - r2)) * (pixel_value - r2) + s2
```

```python
# Define parameters.
r1 = 100
s1 = 0
r2 = 230
s2 = 255


contrast_streched_image = np.vectorize(contrast_streching)(r1, s1, r2, s2, image)
plt.imshow(contrast_streched_image);
cv2.imwrite('contrast_house.jpg', contrast_streched_image)
```

```python
cv2.imwrite('contrast_table.jpg', contrast_streched_image)
```

### Making Background image white

```python
image = cv2.imread('../images/house.jpg')

r1 = 100
s1 = 0
r2 = 120
s2 = 255

contrast_streched_image = np.vectorize(contrast_streching)(r1, s1, r2, s2, image)
plt.imshow(contrast_streched_image)
```
