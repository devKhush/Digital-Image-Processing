```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
image = cv2.imread('../images/sceneary.jpg')
```

```python
plt.imshow(image);
```

```python
a = 110
b = 190
scale = 255

intensity_slicing_img = np.zeros(image.shape, dtype=np.uint8)

for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        for color in [0,1,2]:
            if a <= image[i,j,color] <= b:
                intensity_slicing_img[i,j,color] = scale
            else:
                intensity_slicing_img[i,j,color] = image[i,j,color]
                
plt.imshow(intensity_slicing_img);
```

```python
cv2.imwrite('intensity_sliced_rectangular_2_2.jpg', intensity_slicing_img)
```
