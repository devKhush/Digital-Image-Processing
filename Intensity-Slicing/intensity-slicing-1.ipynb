```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
image = cv2.imread('../images/table.jpg')
```

```python
plt.imshow(image);
```

```python
def intensity_slicing_1(a, b, scaled, pixel_value):
    if 0 <= pixel_value and pixel_value < a:
        return 0
    elif a <= pixel_value <= b:
        return scaled
    else:
        return 0
```

```python
a = 100
b = 230
scale = 255

intensity_slicing_img = np.vectorize(intensity_slicing_1)(a, b, scale, image)
plt.imshow(intensity_slicing_img);
```

```python
cv2.imwrite('intensity_sliced_rectangular_1_2.jpg', intensity_slicing_img)
```
