```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('../images/sceneary.png')
```

```python
plt.imshow(image);
```

```python
bit_images = np.empty(shape=8, dtype='object')
bit_images
```

```python
for bit in range(8):
    ith_bit_image = ((image >> bit) & 1) << bit
    
    print(f"Bit-{bit+1} Image")
    plt.imshow(ith_bit_image)
    plt.show()
    
    bit_images[bit] = ith_bit_image
    
    cv2.imwrite(f'bit_{bit+1}_image.jpg', ith_bit_image)
```

```python
most_informative_img = bit_images[4] + bit_images[5] + bit_images[6] + bit_images[7]

plt.imshow(most_informative_img);
cv2.imwrite('nearly_recovered_image.jpg', most_informative_img)
```
